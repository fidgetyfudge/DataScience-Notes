## Subsets!
> A set A is a subset of B if every element of A is also an element of B.

- In python we use A.issubset(B)
- Every set is a subset of itself!

## Proper Subset!
- A is a proper subset of B iff A is a subset of B and A is not equal to B.


```python
# Subset
A = {1,2}
B = {1,2,3,4}

print(A.issubset(B)) # True!
print(A <= B) # True

# Proper Subset!
print(A < B) # True
print(A < A) # False
```

## Power Sets!

- It is a set of all the subsets of a set!
- Number of subsets for a set of cardinality n = 2^n

### Vacous Truth!
- Empty set is a subset of all sets!
      > Every element of the first set belongs to the second set and it never violates it!

```python
# Method 1 Numbers!
from itertools import combinations

A = [1,2,3]
power_set = []

for r in range(len(A) + 1):
    power_set.extend(combinations(A,r))
print(power_set)

# Method 2 Numbers!
A = {1,2,3}
P = []

for r in range(len(A)+1):
    for subset in combinations(A,r):
        P.append(set(subset))

print(P)

# For letters
A = ["a","b","c"]
power_set = []
for r in range(len(A) + 1):
    for subset in combinations(A,r):
        power_set.append(subset)
print(power_set)
```

```python
# This is how combinations works!
A = {1,2,3,4,5}
B = combinations(A,2)
for i in B:
    print(i)
# This Append vs Extend!
numbers = [1, 2]
new_numbers = [3, 4]

numbers.append(new_numbers)
print(numbers)
print(len(numbers))

numbers = [1, 2]
new_numbers = [3, 4]

numbers.extend(new_numbers)
print(numbers)
print(len(numbers))
```

```python
# Define a function which would do the same!

def power_set(elements):
    subsets = [[]]
    for item in elements:
        new_subsets = []

        for subset in subsets:
            new_subsets.append(subset + [item])

        subsets.extend(new_subsets)

    return subsets

A = [1,2,3]
print(power_set(A))
```
