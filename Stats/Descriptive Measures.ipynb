# Descriptive Measures

> Descriptive measures (or descriptive statistics) are numerical values used to summarize, describe, and understand a dataset.

### This Helps us answer
- What is a typical value?
- How spread out are the values?
- Is the data symmetric or skewed?
- Are there unusual values?

## This includes
1. Measures of Central Tendency
   > These are measures that indicate the most typical values of a data set
2. Measures of Dispersion
   > These measrures indicate the variability of a dataset (How scattered is the data)
3. Measures of Position
   > Tell us where an observation lies relative to others.
4. Measures of Shape
   > Describe the overall form of the distribution.

## Measure of Central Tendency

> A single value that attempts to describe a set of data by identifying the central position within the set of data.

### Mean/ Arithmetic Average for single data
> It is the sum of the observations divided by the nubmer of observations!

- Sample Mean (n)
The mean of raw data is calculated by adding all the data values and dividing by the total number of values.
$$\bar{x} = \dfrac{\sum x}{n}$$

- Population Mean (N)
The mean of a population is the average of all the values in an entire population
$$\mu = \dfrac{\sum x}{N}$$

### Mean of Discrete Grouped Data:
- $x_i$ is the value and then we multiply by frequency in tabel
$$[ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} f_i x_i]$$

### Mean of Continous Grouped Data:
- $x_i$ is the midpoint of the bins
$$
x_i = \frac{\text{Lower Limit} + \text{Upper Limit}}{2}
$$

$$
\bar{x} = \frac{\sum f_i x_i}{\sum f_i}
$$

## Advantages
- Uses Every Observation
- Unique Value for a dataset
- Easy for Mathematical Analysis
- Sum of deaviations is Zero

## Disadvantages

- Sensitive to Outliers
- Not good when data is skewed
- We can't define for Qualitative Data

## Types
1. Arithmatic Mean
2. Weighted Mean


## Sensitivity of Mean
- When each value is increased/decreased by some constant 'c' then the mean is also increased/decreased by 'c'
- When each value is multiplied/divided by some non-zero constant 'c' than the mean is also multiplied/divided by 'c'.

````python
# Mean for frequency table 

def frequency_mean(values, frequencies):

    fx_sum = 0
    f_sum = 0

    for i in range(len(values)):
        fx_sum += values[i] * frequencies[i]
        f_sum += frequencies[i]

    return fx_sum / f_sum
values = [10, 20, 30]
frequencies = [2, 3, 4]

frequency_mean(values, frequencies)
````

````python
# For binned data!
def grouped_mean(intervals, frequencies):

    fm_sum = 0
    f_sum = 0

    for i in range(len(intervals)):

        lower = intervals[i][0]
        upper = intervals[i][1]

        midpoint = (lower + upper) / 2

        fm_sum += midpoint * frequencies[i]
        f_sum += frequencies[i]

    return fm_sum / f_sum
intervals = [[0, 10],[10, 20],[20, 30]]
frequencies = [5, 8, 7]

grouped_mean(intervals, frequencies)
````

## Mode
> It is the observation with the highest frequency.

- Sometimes Mode makes more sense
- Data of shoe size sold [7, 8, 8, 8, 9, 10]
- 8.33 as a mean doesn't have any concrete meaning
- Mode = 8 gives us more information

## Types
1. Unimodal: One Mode [1, 2, 2, 3, 4]
2. Bimodal: Two Mode [1, 2, 2, 3, 3, 4]
3. Multimodal: Many Modes [1, 1, 2, 2, 3, 3]
4. No mode: Every value appears equally often [1, 2, 3, 4, 5]

## Advantage
- Works for Categorical Values
- Not affected by Outliers

## Disadvantage
- May not exist
- May not be unique
- Only cares for frequency

## Sensitive Teset
- Add a constant : New Mode = Old Mode + k
- Multiply a constant: New Mode = Old Mode * k

````python
# Finding Sinlge Mode

data = [2, 3, 4, 4, 4, 5, 6]
freq = {}

for value in data:
    freq[value] = freq.get(value, 0) + 1

max_frequency = max(freq.values())

for key, value in freq.items():
    if value == max_frequency:
        print(key)
````

````python
# Finding Multiple Modes

data = [1, 2, 2, 3, 3, 4]
freq = {}

for value in data:
    freq[value] = freq.get(value, 0) + 1

highest = max(freq.values())

modes = []

for key, value in freq.items():
    if value == highest:
        modes.append(key)

print(modes)
````

## Measure of Dispersion!

> The terms variability, spread, and dispersion are synonyms. They refer to how varied data are in a data set or how spread out the distribution of the data is.

## Why do we need Dispersion?

- Because two datasets can have exactly the same mean, median, and mode while being completely different.
- Dataset 1: [48, 49, 50, 51, 52]
- Dataset 2: [0, 25, 50, 75, 100]
- For both mean and median is 50 and mode doesn't exist but these two dataset are different!
- So we know what is at center now we will know "How far are observations from the center?"



## Range

> Range is the difference between the largest and smallest observation.
> It is the total spread of the data


## Disadvantage
- Sensitive to the outliers
- Ignores most data


## Sesitivity Test

- if k is added in all observations: New Range = Old Range
- if k is multiplied in all obervations: New Range = k * Old Range


````python
datasetA =  [48, 49, 50, 51, 52]
datasetB =  [0, 25, 50, 75, 100]

dfA = pd.DataFrame(datasetA, columns=["Data A"])
dfB = pd.DataFrame(datasetB, columns=["Data B"])

# Tryin tuple unpacking this time!
# And Accessing the dataframe!
for (name,df,col) in [("A", dfA, "Data A"), ("B", dfB, "Data B")]:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode().tolist()
    print(f"Dataset {name} -> Mean: {mean_val}, Median: {median_val}, Mode: {mode_val}")


# range_value = max(datasetA) - min(datasetA)
# range_value1 = max(datasetB) - min(datasetB)

for name, df, col in [("A", dfA, "Data A"),("B", dfB, "Data B")]:
    range_val = max(df[col]) - min(df[col])
    print(f"Dataset {name} -> Range: {range_val}")
````

## Deviations

- Our previous measures incorporated only some of the data values. One way to include all of the data is to compare how far away each piece of data 'x' from 'v' and we define deviation as `x-v` and `|x-v|` would be absolute deviation form!


- This is a great way to understand it! We see that deviations from mean is 0.
- Mean is like a balaced point where some values are above it and some are below it
- Avg deviation from the mean is always 0.
- The sum of the absolute deviations can be considered the total distance our observations are from our central value so it won't cancel!

## MAD
- The total distance is kinda governed by the size of sample too.
- We can define Mean Absolute Deviations as the average distance the various data values are from the central value.
- $\text{MAD}=\frac{1}{n}\sum|x_i-v| \nonumber$
- MAD is minimized for the Median as central tendency.


## Variance:


- As absolute deviation, distance formula is another way for us to cancel the negatives!
- Think of your whole data set as a "point" in some high-dimensional space, and the mean as another "point." The "distance" between them (in this squared-differences sense) gives you a natural way to measure overall spread — just like ordinary distance measures how far apart two points are.
- $\displaystyle d=\sqrt{(x_1-v)^2+(x_2-v)^2+\dots+(x_n-v)^2}\nonumber$



- Note that the median doesn't give us the lowest value for d
- That is why we prefer the Mean and define the Population Variance as:
- Population Variance: $\sigma^2=\frac{\sum (x-\mu )^2}{N}\nonumber$
- Sample Variance: $s^2=\frac{\sum (x-\bar{x})^2}{n-1}\nonumber$
```
The sample mean is always the value that makes the sum of squared deviations as small as possible for that specific sample.
This means when you compute deviations from x(bar) that is calculated from the same sample instead of Population Mean(μ)
you tend to get a sum of squared deviations that's a little too small which it underestimates the true population variance. 
This is called bias. So we take that into consideration and hence the formula change!
```
### Advantage
```
A = [1, 49, 50, 51, 99]
B = [1, 25, 50, 75, 99]

Both has the mean = 50 and Range = 90 so Range doesn't help!
```
- It uses every observation
- Measures ovarall spread
- Has some beautiful properties like V(X) + V(Y) = V(X+Y)

### Disadvantage
- Units becomes squared!
- Sensitive to outliers
- Hard to interpret directly
- Can't define for qualitative data

### Sensitivity Test
- If we add/subtract a constant k to all observations
```
    New Variance = Old Variance 
```
- If we multiply/divide a non-zero constant k to all observations
``` 
    New Variance = K^2 Old Variance (for multiply)
    New Variance = 1/K^2 Old Variance (for division)
```

# Standard Deviation!

> The standard deviation is simply the square root of the variance.
> The population standard deviation is the square root of the population variance.
> The sample standard deviation is the square root of the sample variance.



- We have defined it like this as it makes the units on the measure the same as those of the original data like `cm` and not `cm^2`.

### Sensitivity Test
- If we add/subtract a constant k to all observations
```
    New Variance = Old Variance 
```
- If we multiply/divide a non-zero constant k to all observations
``` 
    New Variance = K Old Variance (for multiply)
    New Variance = 1/K Old Variance (for division)
```

````python
# Writing Python Code for Variance!

data = [2, 4, 6, 8, 10]
mean = sum(data) / len(data)
squared_differences = []

for value in data:
    difference = value - mean
    squared = difference ** 2
    squared_differences.append(squared)

variance = sum(squared_differences) / len(data)

print(variance)
````

## Percentiles

> It answers "What percentage of observations lie below a particular value?"

 > The value of p-percentile is The value below which p% of observations lie.

### Why do we need this man?
```
Say, Student A scored 85 in an easy exam and Student B scored 85 in an extremely difficult exam.
Student A → 60th percentile
Student B → 95th percentile
Student B performed much better relative to peers.
Even though percentage is the same!
```

## Quartiles and the Five number summery

- Quartiles: The three quartiles (Q1, Q2, Q3) divide the data set into four equal parts.
- Q1 (First Quartile): About 25% of the data falls on or below Q1.
- Q2 (Second Quartile): Also known as the median; about 50% of the data falls on or below Q2.
- Q3 (Third Quartile): About 75% of the data falls on or below Q3.
- Q0 : Mininum
- Q4: Maximum
- IQR = Q3 - Q1 (It is the middle 50% spread of the data)
- Range = Q4 - Q0 (It is the total spread of the data)

````python
# Using numpy for 5 number generalization

data = [12, 15, 14, 10, 18, 90, 13, 16, 11, 17, 14, 15, 13, 16, 12, 19, 15, 14, 13, 18]

Q0 = np.percentile(data, 0)
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)   
Q3 = np.percentile(data, 75)
Q4 = np.percentile(data, 0)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q0: {Q0}, Q1: {Q1}, Median: {Q2}, Q3: {Q3}, Q4: {Q4}")
print(f"IQR: {IQR}")
print(f"Outlier bounds: [{lower_bound}, {upper_bound}]")
print(f"Outliers: {outliers}")
````

````python
# Manual Implimentation!
import math
# 1. We sort the data
# we have p percentile [0,1] and size n and we can calulate the np

def percentile(data, p):
    n = len(data)
    data = sorted(data)
    np = n * p
    
    if p == 0:
        return data[0]

    if p == 1:
        return data[-1]
    
    if np.is_integer():
        np = int(np - 1)
        # Note we use slight different formula as how index works
        return (
            data[np - 1]
            + data[np]
        ) / 2

    else:
        np = math.ceil(np)
        return data[np - 1]

data = [35,38,47,58,61,66,68,70,79]
# This was used in the lectures
print(percentile(data, 0))
print(percentile(data, 0.1))
print(percentile(data, 0.25))
print(percentile(data, 0.5))
print(percentile(data, 0.75))
print(percentile(data, 1))
````

## Plotting the percentiles and IQR

## Box-and-Whisker Plot

```
Step 1: Sort all observations in ascending order.
Step 2: Find the Median (Q2)
Step 3: Find the First Quartile (Q1)
> Find the median of the lower half of the data.
Step 4: Find the Third Quartile (Q3)
> Find the median of the upper half of the data.
Step 5: Calculate the Interquartile Range (IQR)
IQR = Q3 − Q1
Step 6: Determine the Outlier Fences+
> Lower Fence = Q1 − 1.5 × IQR
> Upper Fence = Q3 + 1.5 × IQR

Step 7: Identify Outliers
> Less than the Lower Fence, or
> Greater than the Upper Fence
These observations are classified as an outlier.

Step 8: Determine the Whiskers
Lower Whisker: Smallest non-outlier observation.
Upper Whisker: Largest non-outlier observation.

Step 9: Draw a box from Q1 to Q3.
Step 10: Draw a vertical line inside the box at the median (Q2).
Step 11: Draw the Whiskers 
Q1 to the Lower Whisker
Q3 to the Upper Whisker

```

````python
import pandas as pd

data = [12, 15, 14, 10, 18, 90, 13, 16, 11, 17, 14, 15, 13, 16, 12, 19, 15, 14, 13, 18]
s = pd.Series(data)

# --- Central Tendency ---
mean   = s.mean()
median = s.median()
mode   = s.mode()[0]

# --- Spread ---
variance   = s.var()   # sample variance (ddof=1 by default)
std_dev    = s.std()   # sample std dev (ddof=1 by default)
data_range = s.max() - s.min()
Q1  = s.quantile(0.25)
Q2  = s.quantile(0.50)
Q3  = s.quantile(0.75)
IQR = Q3 - Q1

print("--- Central Tendency ---")
print(f"Mean   : {mean}")
print(f"Median : {median}")
print(f"Mode   : {mode}")

print("\n--- Spread ---")
print(f"Variance : {variance}")
print(f"Std Dev  : {std_dev}")
print(f"Range    : {data_range}")
print(f"Q1 / Q2 / Q3 : {Q1} / {Q2} / {Q3}")
print(f"IQR      : {IQR}")

# --- Quick built-in summary ---
print("\n--- s.describe() ---")
print(s.describe())
````
