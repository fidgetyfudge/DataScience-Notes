# Qualitative Datatypes and their description

## Important Terms!
> Statistics is the science of collecting, organizing, summarizing, analyzing, and interpreting data in order to make decisions or draw conclusions.
- *Population* : A population is the complete collection of observations, or measurements of interest.
- *Sample*: A sample is a subset of a population selected for study
  - A good sample is as representative of the population, random and unbiased!
- **Descriptive Statistics**: Descriptive statistics consists of methods used to organize, summarize, and describe the available data.
  - We can describe data with
    1. Measure of Central Tendency
    2. Measure of Dispersion
    3. Tables
    4. Charts 
- **Infrential Statistics**: Inferential statistics consists of methods used to make conclusions, predictions, or generalizations about a population using sample data.
  - We talk about the Hypothesis Testing,Confidence Intervals,Regression,Correlation,Estimation

## Data
**Data**: Data: Raw facts, measurements, observations, or values collected for analysis.
**Dataset**: Dataset: An organized collection of related data, usually arranged in rows and columns for analysis.
- These data are are majorly two types
  1. Structured Data
  2. Unstructured Data

## Concept of variability!

In statistics, data are collected about variables. Data is the plural term used to refer to multiple bits of information together. The term data set can also be used to refer to data from one or many variables. Datum is the singular form of data which is used to refer to one piece of information. Variability is a broad term that refers to the fact and ways in which things can differ. A variable is anything that is measured which is not always the same. A constant is anything that is measured and always is the same!

## What is a variable?
> It is a characteristic/attribute that can have any values and we collect the data for these variables!
- We have classified our datatypes/varaibles into Qualitative and Quantitative
- We also have classified our data in Structured Data and Unstructured Data

### Operationalize a varibale!
>Operationalization is the process of defining an abstract concept as a measurable variable by specifying exactly how it will be observed, recorded, or quantified.
- Example, We can't js measure intelligence but IQ boys!!! which allows to define a varable and gather data and do research.

## Types of Variables!
1. Qualitative Variables
> Qualitative variables vary in: [Characteristic,Category,Type,Kind]
- Examples Eye Color,Blood Group, Gender, College Major.
- A Physics major varies from a Chem major in type
- We can't say if one is greater than the other
- They are more like labels!
###  Subtypes:
- 1. Nominal: No natural order exists
  2. Ordinal: Natural order exists

2. Quantitative Variables
> Quantitative variables vary in amount
- Examples: Weight (kg),Height (cm),Income(Rs)
### Subtypes:
- 1. Discrete/Discontinous Variable: Can Take only discrete values! Ex- Number of children in a family
- 2. Continous Variable: Can take any values.. Ex: Hours of sleep one took
 
## Scale of Measurement!
### Nominal Scale
>A scale of measurement in which observations are classified into categories without any natural ordering.
- We can talk about frequency(relative), and mode.
- Example: Comfortable temperature, Uncomfortable temperature
### Ordinal Scale:
>A scale of measurement in which observations can be ranked or ordered, but the differences between ranks are not necessarily equal or known.
- Frequency, relative frequency, mode and median makes sense!
- Example: Very warm, warm, cold, very cold
### Interval Scale
>A quantitative scale with equal intervals between values but no true zero point.
- Addition, Subtraction Makes sense but no ratios exists
- Example:Temperature in Celsius and Farenheit
- 20 degrees is not twice as hot as 10 degrees!
### Ratio Scale
>A quantitative scale with equal intervals and a true zero representing complete absence of the quantity being measured.

- All Arithmetic Operations makes sense!
- Example: Temperature in Kelvin
## Independent and Dependent Variable!

- Univariate Statistics:
  > Univariate statistics are statistical procedures and results that involve the analysis or summary of one variable at a time.
  - Example: Marks: We can analyse Mean, Median and Mode of it// Distribution of Marks

- Bivariate Statistics:
  > Bivariate statistics are statistical procedures and results that involve two variables and the relationship or association between them.
  - Example: Study Hours and Grades: We can ask "Do students who study more tend to score higher marks?"

- Multivariate Statistics:
  > Multivariate statistics are statistical procedures and results that involve three or more variables simultaneously.
  - Example: Variables[Marks, Sleep Hours, Attendence, Study hours], We can ask "How do study hours, attendance, and sleep affect exam marks?"

- Independent Variable!
  > An independent variable is the variable that is assumed or hypothesized to be the cause in a cause-effect relationship.
  - Example: Does studying increases grades? Studying Hours is dependent variable bc because studying is assumed to influence grades.

- Dependent Variable!
  > A dependent variable is the variable that is assumed or hypothesized to be affected by the independent variable in a cause-effect relationship.
  - Example: Does studying increase grades? Grades are dependent variable bc because grades are assumed to depend on the amount of studying.
```Note: The same variable can be independent in one study and dependent in another. ```


## Graphical Methods of Qualitative Data
## Frequency Table
> A frequency table is a table that shows how many times each value or category occurs in a dataset.

The count of occurrences is called the frequency.
They are the basis of bar graphs and pie charts!

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = ["A", "B", "A", "A", "B", "C", "A", "D", "C", "B"]

# Creating the spreadsheet
df = pd.DataFrame(data, columns = ["Category"])
# Accessing and counting!
frequency_table = df["Category"].value_counts()

print(frequency_table)

````

````python
# We can try to achieve the same in regular python way

data = ["A", "B", "A", "A", "B", "C", "A", "D", "C", "B"]
frequency_dictionary = {}

for category in data:
    if category not in frequency_dictionary:
        frequency_dictionary[category] = 1
    else:
        frequency_dictionary[category] += 1
print(frequency_dictionary)   
````

## Relative Frequency and better organisation

- We will now understand relative frequwncy which is like what porportion of all does belong to a category!

````python
# Pure Python Way Relative and Percentage Frequency!
frequency_dictionary = {'A': 4, 'B': 3, 'C': 2, 'D': 1}

total = len(data)
for category in frequency_dictionary:
    relative_frequency = frequency_dictionary[category]/ total
    percentage_frequency = relative_frequency*100
    print(f"{category} has a relative frequency of {relative_frequency} // {percentage_frequency}%")
````

````python
# Pandas have a built in command here as well

frequency = df["Category"].value_counts()

# relative_frequency is a series btw
relative_frequency = (
    df["Category"].value_counts(normalize=True)
)

percentage_frequency = (
    relative_frequency * 100
)

table = pd.DataFrame({
    "Frequency": frequency,
    "Relative Frequency": relative_frequency,
    "Percentage Frequency": percentage_frequency
})

print(table)
````

## Pi Chart!
    - It is representation of relative frequency as the sector of a circle.
    

````python
# Plotting Pie Chart

Animals = ["Dog", "Cat", "Bird", "Fish"]
sizes = [4, 3, 2, 1]

plt.pie(sizes)
plt.show()
````

````python
# Improving!

plt.pie(
    sizes,
    labels = Animals,
    autopct= "%1.2f%%"

)
# Adding Some labels
plt.title("Pet Distribution")

plt.show()
````

````python
# More Styling!

plt.pie(
    sizes,
    labels = Animals,
    autopct= "%1.0f%%",
    startangle = 90, # Where u start the category
    shadow = True, # Shadows Duh
    explode = [0,0,0.2,0] # Make a slice pop out
)
# Adding Some labels
plt.title("Getting Better Design!")
plt.show()
````

- We will Now Try To Import it from the Pandas Library!


````python
data = [
    "Dog","Cat","Dog","Bird","Dog","Cat","Bird","Dog"
]

df = pd.DataFrame(
    data,
    columns=["Pets"]
)
frequency = df["Pets"].value_counts()

plt.pie(
    frequency.values,
    labels = frequency.index,
    autopct = "%1.1f%%",
    # colors = ["red", "green", "orange"],
    colors=plt.cm.Set3.colors, # Adds Auto Color
    wedgeprops={
        # Adds the border!
        "edgecolor": "black"
    }
)
centre_circle = plt.Circle(
    (0, 0),
    0.45,
    fc="white"
)
plt.title("Hmmmmmmm")
plt.gca().add_artist(centre_circle)
plt.show()
````

## Misleading Data// Violation of data!

Area Principle: The area occupied by a graphical element should be proportional to the value it represents. Violating this can exaggerate or understate differences.

Truncated Graph: A graph where the axis (usually the y-axis in a bar chart) does not start at zero, making differences appear larger than they actually are.

Round-Off Error: Error introduced when values are rounded. Rounded values may not add up exactly to the original total, causing slight discrepancies.

## Descriptive Methods
- Mode: How frequent is one category? (Can be talked for either nominal and ordinal data)
- Median: For ordinal variable it is the category of the middle observation of sorted data

````python
ratings = ["Poor","Average","Good","Good","Excellent","Good"]

# Finding the Mode!
frequency = {}
for rating in ratings:
    frequency[rating] = frequency.get(rating, 0) + 1
    #Gimme the value associated with color and if doesn't exist yet, pretend its value is 0
print(frequency)

````

````python
# Mode!
mode = max(frequency, key=frequency.get)

print(mode)
````

````python
ratings = [
    "Poor",
    "Average",
    "Good",
    "Good",
    "Excellent"
]
# We can only define it on the data which has some order!
order = {
    "Poor": 1,
    "Average": 2,
    "Good": 3,
    "Excellent": 4
}

# This function receives one rating
# and returns its rank/position
def get_rank(rating):
    return order[rating]

sorted_ratings = sorted(
    ratings,
    key=get_rank
)
# sorted(ratings, key=get_rank) is basically doing: get_rank("Poor") for each rating
# Kind of like a loop which sorts our ratings! 

# Find middle position
middle_index = len(sorted_ratings) // 2

# Pick middle item
median = sorted_ratings[middle_index]

print("Median:", median)
````
