```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset
df = pd.DataFrame({
    "Marks": [35, 38, 47, 58, 61, 66, 68, 70, 79]
})

s = df["Marks"]

print("=" * 50)
print("DESCRIPTIVE STATISTICS")
print("=" * 50)

# Central Tendency
print("Mean:", s.mean())
print("Median:", s.median())
print("Mode:")
print(s.mode().to_list())

print()

# Dispersion
print("Minimum:", s.min())
print("Maximum:", s.max())
print("Range:", s.max() - s.min())

print()

print("Sample Variance:", s.var())
print("Population Variance:", s.var(ddof=0))

print()

print("Sample Std Dev:", s.std())
print("Population Std Dev:", s.std(ddof=0))

print()

# Percentiles & Quartiles
print("Q1 (25th Percentile):", s.quantile(0.25))
print("Q2 (Median):", s.quantile(0.50))
print("Q3 (75th Percentile):", s.quantile(0.75))

print()

print("10th Percentile:", s.quantile(0.10))
print("90th Percentile:", s.quantile(0.90))

print()

# IQR
q1 = s.quantile(0.25)
q3 = s.quantile(0.75)

iqr = q3 - q1

print("IQR:", iqr)

print()

# Outliers
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("Lower Fence:", lower)
print("Upper Fence:", upper)

outliers = s[(s < lower) | (s > upper)]

print("Outliers:")
print(outliers)

print()

# Additional Statistics
print("Count:", s.count())
print("Sum:", s.sum())
print("Unique Values:", s.nunique())
print("Skewness:", s.skew())
print("Kurtosis:", s.kurt())

print()

print("=" * 50)
print("DESCRIBE")
print("=" * 50)

print(s.describe())

# -------------------------
# VISUALIZATIONS
# -------------------------

# Histogram
s.hist(bins=5)
plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Box Plot
s.plot.box()
plt.title("Box Plot")
plt.show()

# Frequency Bar Chart
s.value_counts().sort_index().plot.bar()
plt.title("Bar Chart")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
s.value_counts().sort_index().plot.pie(
    autopct="%1.1f%%"
)
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```
