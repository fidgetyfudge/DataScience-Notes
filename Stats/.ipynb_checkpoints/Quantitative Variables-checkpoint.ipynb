# Numerical Varaibles

## What is it?
> Numerical (or quantitative) data are data whose values are expressed as numbers and for which arithmetic operations such as addition, subtraction, averaging, and comparison are meaningful.

## Types
1. Discrete
 > Discrete data arise from counting and can take only specific, distinct values.
- We get this through counting!
- Example:
Number of books on a shelf,Number of goals scored,Number of children in a family

2. Continous
 > Continuous data arise from measurement and can take any value within an interval.
- We obtain this through measurents!
- Example:
Height,Weight,Temperature


## Graphing the Discrete Data
- We will treat each distinct values of the discrete data as a "Categorical Data" for small number of distinct values.
- We can then create frequency tables, bar graph and pie chart

### Example
- For our `data = [2,1,3,4,5,2,3,3,3,4,4,1,2,3,4]` which represent the number of people in a household.
- We have 1,2,3,4 and 5 as distinct values and we will consider them as categorical data to plot the graphs!

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Constructing the Frequency Table!

# This represents the number of people in a household!
data = [2,1,3,4,5,2,3,3,3,4,4,1,2,3,4]

df = pd.DataFrame(data, columns=["Number of People"])

frequency = (df["Number of People"].value_counts().sort_index())

relative_frequency = (
    df["Number of People"]
    .value_counts(normalize=True)
    .sort_index()
)
percentage = relative_frequency * 100

frequency_table = pd.DataFrame({
    "Frequency": frequency,
    "Relative Frequency": relative_frequency,
    "Percentage": percentage
})

print(frequency_table)

````

````python
# Cosntructing the Bar Graph!
frequency.plot(kind="bar")

plt.title("Number of People in Household")
plt.xlabel("Number of People")
plt.ylabel("Frequency")

plt.show()
````

````python
# Graphing the Pie Chart!

plt.pie(
    frequency.values,
    labels = frequency.index,
    autopct = "%1.1f%%"
)
plt.title("Household Size Distribution")
plt.axis("equal")
plt.show()
````

## Graphing Continous Data

### Organizing the continous data!
- We can have many distinct values if we follow the same logic!
- We will try organize the data in a number of classes!

### Guidelines of creating classes! (Bins)
- Number of classes: Don't create way too many classes!
- Each observation should belong to a class and no observation should belong to more class
- Try making class interval of equal length!

### Some new terms
- Lower class limit: The smallest value that could go in a class.
- Upper class limit: The largest value that could go in a class.
- Class width: The difference between the lower limit of a class and the lower limit of next-higher class.
- Class mark: The average of the two class limits of a class.
- Modal Class: The interval with the highest frequency!
**A Class interval contains its left-ebd but not its right-end boundary point**
- Example: [30,40) 30 belongs but not 40:
- And hence only marks that is: `30 <= mark < 40` will belong!
  

````python
# Tables for grouped continous data!

import pandas as pd

marks = [
    68, 79, 38, 68, 35, 70, 61, 47, 58, 66,
    60, 45, 61, 60, 59, 45, 39, 80, 59, 62,
    49, 76, 54, 60, 53, 55, 62, 58, 67, 55,
    86, 56, 63, 64, 67, 50, 51, 78, 56, 62,
    57, 69, 58, 52, 42, 66, 42, 56, 58
]

frequency = {
    "30-40": 0,
    "40-50": 0,
    "50-60": 0,
    "60-70": 0,
    "70-80": 0,
    "80-90": 0
}

for mark in marks:
    if 30 <= mark < 40:
        frequency["30-40"] += 1
    elif 40 <= mark < 50:
        frequency["40-50"] += 1
    elif 50 <= mark < 60:
        frequency["50-60"] += 1
    elif 60 <= mark < 70:
        frequency["60-70"] += 1
    elif 70 <= mark < 80:
        frequency["70-80"] += 1
    elif 80 <= mark < 90:
        frequency["80-90"] += 1

total = len(marks)

table = pd.DataFrame({
    "Frequency": frequency.values(),
    "Relative Frequency": [f / total for f in frequency.values()],
    "Percentage Frequency": [(f / total)*100 for f in frequency.values()]
}, index=frequency.keys())

print(table)
````

````python
# Plotting the Historgrams!

plt.bar(
    frequency.keys(),
    frequency.values()
)
plt.title("Frequency Distribution of Marks")
plt.xlabel("Class Interval")
plt.ylabel("Frequency")
plt.show()
````

## Histograms!

> It is bar chart but for out continous-data!
> It display the distribution of continous numerical data not comparison unlike bar chart.

- The data is divided into bins(clases) and the frequency of observations in each interval is shown by the height fo the bars.
- The bar of histogram touches (continous data)
### Advantages
- It helps us understand:
```
Shape of data, Spread of data, Center of data, Outliers, Skewness, Distribution pattern
```

## Shape of histograms

1. Symmetric Distributions (Balanced)
> `Mean ~ Median`

2. Right Skewed (Positive Skew)
> `Mean > Median`

3. Left Skewed (Negative Skew)
> `Mean < Median`

4. Bimodal Distribution
> It had two Modes!

````python
# Plotting Histograms

plt.hist(marks)
plt.show()

# The bins would be chosen automatically!
````

````python
plt.hist(
    marks,
    bins = [30,40,50,60,70,80,90],
    # bins = 5,
    # bins = 10,
    density= True, # Plots the Relative Frequency Histogram
    color = "skyblue",
    edgecolor = "black",
    orientation = "horizontal"
)
# Low bins - Information lost
# High bins - Noisy
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
````

````python
# Plotting using the pandas

df = pd.DataFrame(
    marks,
    columns =["Marks"]
)
df["Marks"].hist(
    bins = 5
)
plt.show()
````

## Stem and Leaf Diagram
>A stem-and-leaf plot displays quantitative data by splitting each number into:

Stem → leading digit(s)
Leaf → last digit

### Example
[12 14 15 18 21 22 25 29]
```
This is how we would plot it!
1 | 2 4 5 8
2 | 1 2 5 9
```

## Advantage

- You can see every observation.
- Original values are preseved.
- Best for small datasets
- Shapes can still show symettr6y, skewness or clusters

# Disadvantage
- Large datasets
- Difficult with Decimals
- Bigger values like 143535 it is awkward

````python

df = pd.DataFrame(
    {"Marks":[12,14,15,18,21,22,25,29]}
)

df["Stem"] = df["Marks"] // 10
df["Leaf"] = df["Marks"] % 10

print(df)
````
