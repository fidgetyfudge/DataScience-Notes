```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Dataset
data = np.array([35, 38, 47, 58, 61, 66, 68, 70, 79])

print("=" * 50)
print("DESCRIPTIVE STATISTICS")
print("=" * 50)

# Central Tendency
print("Mean:", np.mean(data))
print("Median:", np.median(data))

# Mode
values, counts = np.unique(
    data,
    return_counts=True
)

mode = values[np.argmax(counts)]

print("Mode:", mode)

print()

# Dispersion
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
print("Range:", np.max(data) - np.min(data))

print()

print("Population Variance:", np.var(data))
print("Sample Variance:", np.var(data, ddof=1))

print()

print("Population Std Dev:", np.std(data))
print("Sample Std Dev:", np.std(data, ddof=1))

print()

# Percentiles
print("Q1:", np.percentile(data, 25))
print("Q2:", np.percentile(data, 50))
print("Q3:", np.percentile(data, 75))

print()

print("10th Percentile:", np.percentile(data, 10))
print("90th Percentile:", np.percentile(data, 90))

print()

# IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

iqr = q3 - q1

print("IQR:", iqr)

print()

# Outliers
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("Lower Fence:", lower)
print("Upper Fence:", upper)

outliers = data[
    (data < lower) |
    (data > upper)
]

print("Outliers:", outliers)

print()

# Additional Statistics
print("Count:", data.size)
print("Sum:", np.sum(data))
print("Unique Values:", len(np.unique(data)))

print()

# Five Number Summary
print("Five Number Summary")
print("Min:", np.min(data))
print("Q1:", np.percentile(data, 25))
print("Median:", np.percentile(data, 50))
print("Q3:", np.percentile(data, 75))
print("Max:", np.max(data))

# -------------------------
# VISUALIZATIONS
# -------------------------

# Histogram
plt.hist(data, bins=5)
plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.boxplot(data)
plt.title("Box Plot")
plt.show()

# Bar Chart
unique, counts = np.unique(
    data,
    return_counts=True
)

plt.bar(unique, counts)
plt.title("Bar Chart")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
plt.pie(
    counts,
    labels=unique,
    autopct="%1.1f%%"
)

plt.title("Pie Chart")
plt.show()
```
